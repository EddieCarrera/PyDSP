# Convolution

Convolution is important because it related three signals of interest: the input, output, and impulse response. This chapter presents two different viewpoints; the input side algorithm and output side algorithm.

## Convolution & Systems



If we know a system's impulse response, we can calculate what the output will be for any possible input. This means we know _everything_ about the system. 

The impulse response is called __filter kernel__ if the system is considered a filter. In image processing, the impulse response is called the __point spread function__.

The output of a convolution will always be:
$$ N+M-1 $$

It's important to remember that <ins>the impulse response is not fully immersed in the input signal</ins>. If the impulse response is M
points in length, the first and last M-1 samples in the output signal are based on less information than the samples between. In other words, this these points are what makes the first and last computations of the convolution. This is analogous to an electronic circuit requiring a certain amount of time to stabilize after the power is applied. The difference is that this transient is easy to ignore in electronics, but very prominent in DSP.

## The Convolution Summation

The convolution sum equation for is:

$$ y[n] = x[n] \ast h[n]=\sum_{j=0}^{M-1} h[j]x[n-j]$$

The index, i, determines which sample in the output signal is being
calculated, and therefore corresponds to the left-right position of the convolution machine. In computer programs performing convolution, a loop makes this index run through each sample in the output signal. To calculate one of the output samples, the index, j, is used inside of the convolution machine. As j runs through 0 to M-1, each sample in the impulse response, h[ j ], is multiplied by the proper sample from the input signal, x[ i-j ]. All these products are added to produce the output sample being calculated.
